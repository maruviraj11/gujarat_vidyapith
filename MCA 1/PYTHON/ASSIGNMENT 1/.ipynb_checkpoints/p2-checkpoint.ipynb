b=int(input("Enter a any number:"))

add=a+b

print(f"a==b->{a==b}")
print(f"a!=b->{a!=b}")

print(f"a<b->{a<b}")

print("------------------Bitwise operation----------------------------------")
print(f"a & b={a & b}")

print("-----------------Membership test--------------------------------------")
predefined_list = [10, 20, 30, 40, 50]
print(f"a in {predefined_list} → {a in predefined_list}")
print(f"b not in {predefined_list} → {b not in predefined_list}")

print("\n--- Identity Test ---")
print(f"a is b → {a is b}")
